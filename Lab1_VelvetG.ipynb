```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Problema 1

```python
def problema1 (x, y):
    img = cv2.imread(x)
    alto = img.shape[0]
    ancho = img.shape[1]
    lienzo = np.zeros((alto, ancho, 3))
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            
            if y == 1:
                lienzo[i, j] = [blue, 0, 0]
            if y == 2:
                lienzo[i, j] = [0, green, 0]
            if y == 3:
                lienzo[i, j] = [0, 0, red]
            if y == 10:
                lienzo[i, j] = [0, green, red]
            if y == 20:
                lienzo[i, j] = [blue, green, 0]
            if y == 30:
                lienzo[i, j] = [blue, 0, red]
            
    cv2.imwrite("problema_1.jpg", lienzo)
    img2 = cv2.imread("problema_1.jpg")
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)        
    plt.imshow(img2)
    plt.show()
```

## Problema 5

### Escala de grises ponderado

---


#### Las imagenes tiene 3 canales RGB (rojo, verde y azul) y la intensidad de cada pixel va de 0 a 255. Convertir una imagen a escala de grises es posible al reducir los tres canales a 1 y uno de los enfoques para realizar esta conversión es el promedio ponderado. Hay opciones estándar de promedios ponderados como el BT.601 y BT.709 en donde cada canal se multiplica por el promedio, generando así una imagen a escala de grises combinando todas las intensidades que genera la variación en gris de cada canal RGB.
###### **Fuente:** Visión y robótica https://visionyrobotica.com/convertir-imagen-rgb-escala-de-grises/ consulta hecha 27 de julio de 2022

```python
### Promedio ponderado BT.601
def problema5(x):
    img = cv2.imread(x)
    alto = img.shape[0]
    ancho = img.shape[1]
    prom_ponderado = np.zeros((alto, ancho, 1))
    for i in range(0, alto):
        for j in range(0, ancho):
            pixel = img[i, j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            prom_ponderado[i, j] = int(0.299*blue + 0.587*green + 0.114*red)
    cv2.imwrite("problema_5_gris_ponderado.jpg", prom_ponderado)
    img2 = cv2.imread("problema_5_gris_ponderado.jpg")
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)        
    plt.imshow(img2)
    plt.show()
```

## Problema 6

### Espacio de color HSV (Hue Saturation Value)

---

#### Se trata de una transfomación no lineal del espacio de color RGB. Esta tiene forma de cono en donde el circulo base representa el matiz de colores como un arcoiris, la saturación que va desde el centro al borde del circulo que representa la instensidad del color y el valor que va desde el pico del cono al centro del circulo representando la intensidad de negro a blanco (0 a 255) que se le agrega al color. 
#### "Es común que deseemos elegir un color adecuado para alguna de nuestras aplicaciones, cuando es así resulta muy útil usar la ruleta de color HSV. En ella el matiz (H) se representa por una región circular; una región triangular separada, puede ser usada para representar la saturación (S) y el valor (V) del color. Normalmente, el eje horizontal del triángulo denota la saturación, mientras que el eje vertical corresponde al valor del color. De este modo, un color puede ser elegido al tomar primero el matiz de una región circular, y después seleccionar la saturación y el valor del color deseados de la región triangular."
